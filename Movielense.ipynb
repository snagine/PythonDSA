# Few examples to understand the joins

```python
data = {'City': ['New York', 'Chicago', 'New York', 'Chicago', 'New York'],
        'Temperature': [10, 20, 15, 25, 12],
        'Humidity': [50, 60, 55, 65, 52]}

df = pd.DataFrame(data)
#print(df)
# Group by 'City' and calculate the mean temperature and humidity for each city
grouped = df.groupby('City').mean()

#print(grouped)

df.groupby('City')['Humidity'].mean()
```

```python
product = pd.DataFrame({
    "Customer_id" : [1,2,3,4,5,6],
    "Product":['Oven', 'Oven', 'Oven', 'Television', 'Television', 'Television']
})

state = pd.DataFrame({
    "Customer_id" : [2,4,6,7,8],
    "State":['California', 'California', 'Texas', 'Las Vegas', 'Las Vegas']
})
print(product)
print(state)
prod_stat_merged = pd.merge(product, state, on='Customer_id')
print(prod_stat_merged)

prod_stat_merged_left = pd.merge(product, state, how = 'left')
print(prod_stat_merged_left)

prod_stat_merged_right = pd.merge(product, state, how = 'right')
print(prod_stat_merged_right)

prod_stat_merged_outer = pd.merge(product, state, how = 'outer')
print(prod_stat_merged_outer)
```
